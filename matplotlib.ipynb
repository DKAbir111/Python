```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import random
```

```python
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])
font1 = {'family':'serif','color':'blue','size':20}
plt.plot(x,y, linestyle = 'dashed', color = 'red')
plt.title("X vs Y", fontdict=font1)
plt.xlabel("X axis", fontdict=font1)
plt.ylabel("Y axis", fontdict=font1)
plt.show()
```

```python
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])
plt.plot(x,y, marker= 'o')
plt.show()
```

```python
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])
plt.plot(x,y, 'o:r')
plt.show()
```

```python
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])
plt.plot(x,y, 'o')
plt.grid(axis='x')
plt.show()
```

```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)
plt.grid()
plt.show()
```

```python
x1= np.array([1,2,3,4])
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.subplot(1,2,1)
plt.plot(x1,y1)
plt.title("plot1")

plt.subplot(1,2,2)
plt.plot(x1,y2)
plt.title("plot2")

plt.suptitle("Plot")
plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x,y,colors, cmap='Wistia')
plt.colorbar()
plt.show()
```
