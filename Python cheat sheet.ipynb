# **Function**

```python
def factorial(n):
  if n==1:
    return 1
  else:
    return n*factorial(n-1)

m=factorial(4)
m
```

# **Class and Object**

```python
class Student:
  def __init__(self,name,roll):
    self.name=name
    self.roll=roll
  def display(self):
    print("Name= ", self.name," ","Roll= ",self.roll)
```

```python
student1=Student("Bithi",15043)
student1.display()
```

```python
print(len(student1.name))
print(student1.roll)
student1.name="Islam"
student1.display()
```

```python
class MSc_Student(Student):
  Dept_name="ICT"
  def __init__(self,name,roll,cgpa):
    Student.__init__(self,name,roll)
    self.cgpa=cgpa
  def set_cgpa(self,cgpa):
    self.cgpa=float(cgpa*2)
  def get_cgpa(self):
    return self.cgpa
```

```python
msc_student1=MSc_Student("Bithi","15043",3.88)
msc_student1.display()
print("CGPA=",msc_student1.get_cgpa())
msc_student1.set_cgpa(3.5)
print("CGPA=", msc_student1.get_cgpa())
print(msc_student1.Dept_name)
```
